## Müşteri Hizmetleri Otomasyonu

Amaç: Bir müşterinin, satın aldığı ürünün garantisi hakkında sorduğu bir soruya, ürünün Sıkça Sorulan Sorular (SSS) sayfasından cevap vermek.

```python
# Gerekli kütüphaneler: pip install sentence-transformers scikit-learn numpy
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# --- BİLGİ KAYNAĞI ---
KNOWLEDGE_BASE = """
Soru: Siparişim ne zaman kargoya verilir?
Cevap: Siparişleriniz, ödeme onayını takiben 2 iş günü içinde anlaşmalı kargo firmamıza teslim edilir. Kargo takip numaranız e-posta adresinize gönderilecektir.

Soru: Ürünlerinizin garanti süresi ne kadardır?
Cevap: Tüm elektronik ürünlerimiz, fatura tarihinden itibaren 2 yıl distribütör garantisi altındadır. Garanti kapsamı, kullanım kılavuzunda belirtilen kullanıcı hatalarını içermez. Kahve makineleri gibi küçük ev aletleri de bu kapsama dahildir.

Soru: Ürün iade politikanız nedir?
Cevap: Satın aldığınız ürünü, ambalajı bozulmamış ve yeniden satılabilir durumda olması koşuluyla, teslim tarihinden itibaren 14 gün içinde koşulsuz iade edebilirsiniz.
"""

# --- MODÜLER FONKSİYONLAR ---
def chunk_text(text: str) -> list[str]:
    return [p.strip() for p in text.split('\n\n') if p.strip()]

def get_embeddings(chunks, model):
    return model.encode(chunks)

def find_most_similar_chunks(query, embeddings, chunks, model, top_k=1):
    query_embedding = model.encode([query])
    similarities = cosine_similarity(query_embedding, embeddings)[0]
    top_k_indices = np.argsort(similarities)[-top_k:][::-1]
    return [chunks[i] for i in top_k_indices]

def generate_answer(query, context):
    prompt = f"Bağlam: {context}\n\nSoru: {query}\n\nCevap:"
    print("--- LLM'e Gönderilen Nihai Prompt ---\n", prompt)
    mock_response = f"SSS sayfamıza göre, '{query}' sorunuzun cevabı şudur: {context}"
    return mock_response

# --- ANA PROGRAM AKIŞI ---
if __name__ == "__main__":
    embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
    user_query = "Yeni aldığım kahve makinesinin garantisi var mı, ne kadar süreyle?"
    
    print("--- Müşteri Hizmetleri RAG Demosu ---")
    print(f"Kullanıcı Sorusu: '{user_query}'\n")

    chunks = chunk_text(KNOWLEDGE_BASE)
    print(f"--- 1. Parçalama Sonucu: {len(chunks)} adet SSS bulundu. ---\n")
    
    document_embeddings = get_embeddings(chunks, embedding_model)
    
    context_chunks = find_most_similar_chunks(user_query, document_embeddings, chunks, embedding_model)
    context_for_llm = "\n".join(context_chunks)
    print("--- 2. Arama Sonucu: En Alakalı SSS Bulundu ---\n", context_for_llm, "\n")
    
    final_answer = generate_answer(user_query, context_for_llm)
    print("\n--- 3. Nihai Cevap Üretildi ---")
    print(final_answer)
```
